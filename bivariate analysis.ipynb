```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from pandas import *
from pylab import * 
import seaborn as sns
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['SimHei'] # 指定默认字体
mpl.rcParams['axes.unicode_minus'] = False   # 解决保存图像是负号'-'显示为方块的问题
```

```python
df = pd.read_csv('plottt.csv', encoding = 'utf8',index_col=0)
```

```python
df.rename(columns={'0':'rating'}, inplace=True)
df.rename(columns={'1':'brand'}, inplace=True)
df.rename(columns={'2':'popularity'}, inplace=True)
df.rename(columns={'3':'types'}, inplace=True)
df.rename(columns={'4':'gender'}, inplace=True)
```

```python
# brand: Loewe 001 Man 檀香木 麝香 香柠檬 香根草 皮革 有四个原料都处于top10 全部都处于top20
#        Agent Provocateur 玫瑰 麝香 藏红花 香根草 雪松 有四个原料处于top10  
#        Elie Saab Le Parfum 橙花 茉莉 蜂蜜 广藿香 玫瑰 有三个原料处于top10   所以它们各种centrality稳定在前五

#        John Varvatos Artisan 克里曼丁红橘 橙花 蜜橘 生姜 马郁兰 
#        John Varvatos Dark Rebel 皮革 烟草 朗姆酒 糖 香草     只有皮革和香草比较常见
#        Vera Wang Princess 香草 黑巧克力 番石榴 睡莲 提亚蕾花  只有香草比较常见
#        Jean Desprez Bal a Versailles 麝猫香 树脂 妥鲁香脂 安息香脂 琥珀  只有琥珀比较常见

# type: 花香甘苔调: 9款香水 茉莉 广藿香 (柠檬 玫瑰)
#       花果香美食调：祖玛珑 杏桃花与蜂蜜 油桃 桃子 绿叶 李子 洋槐 所以各种centrality都比较低
```

```python
df['degreecentrality']=[i/199 for i in df['degree']]
df['weighteddegreecentrality']=[i/199 for i in df['weighted degree']]
```

```python
df=df[['gender','brand','types','rating','popularity','betweenesscentrality','closnesscentrality','degreecentrality','eigencentrality','clustering']]
df.head(1)
```

# pivot_table

```python
tb1 = pivot_table(df, values=['rating','degreecentrality'],
                     index=['brand'],
                     aggfunc={'rating':np.mean,'degreecentrality': np.mean})
tb1 = tb1.sort_values('degreecentrality',ascending=False)
```

# gender, type and brand X rating and popularity

```python
df1 = pivot_table(df, values=['rating','popularity'], index=['gender'],aggfunc={'rating': np.mean,'popularity':np.mean})
plt.figure(figsize=(8,3))
for i in range(1,3):
    plt.subplot(1,2,i)
    df1 = df1.sort_values(df1.columns[i-1],ascending=False)
    plt.bar(df1.index,df1[df1.columns[i-1]],width = 0.3,facecolor = 'lightskyblue')
    plt.title('gender vs. '+ df1.columns[i-1],fontsize=14)
    plt.xticks(fontproperties="SimHei",fontsize=13)
    plt.tight_layout()
    sns.set_style("darkgrid")
```

```python
df2 = pivot_table(df, values=['rating','popularity'], index=['brand'],aggfunc={'rating': np.mean,'popularity':np.sum})
plt.figure(figsize=(12,8))
for i in range(1,3):
    plt.subplot(2,1,i)
    df2 = df2.sort_values(df2.columns[i-1],ascending=False)
    plt.bar(df2.index,df2[df2.columns[i-1]],width = 0.8,facecolor = 'lightskyblue')
    plt.title('brand vs. '+ df2.columns[i-1],fontsize=14)
    plt.xticks(fontproperties="SimHei",fontsize=10,rotation=75)
    plt.tight_layout()
    sns.set_style("darkgrid")
```

```python
df3 = pivot_table(df, values=['rating','popularity'], index=['types'],aggfunc={'rating': np.mean,'popularity':np.mean})
plt.figure(figsize=(12,8))
for i in range(1,3):
    plt.subplot(2,1,i)
    df3 = df3.sort_values(df3.columns[i-1],ascending=False)
    plt.bar(df3.index,df3[df3.columns[i-1]],width = 0.8,facecolor = 'lightskyblue')
    plt.title('type vs. '+ df3.columns[i-1],fontsize=14)
    plt.xticks(fontproperties="SimHei",fontsize=10,rotation=75)
    plt.tight_layout()
    sns.set_style("darkgrid")
```
