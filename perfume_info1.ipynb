```python
import unittest
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd 
browser = webdriver.Chrome("C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe")
```

```python
result_list=[]
pages = [str(i) for i in range(1,11)] 

for page in pages:
    browser.get("https://www.nosetime.com/top200.php?type=trade&page="+page+"#list")
    browser.implicitly_wait(3)
    result = browser.find_elements_by_xpath('//*[@id="top"]/ul/li') 
    mydict={}
    
    for x in range(1,len(result)+1): 
        name_element = browser.find_element_by_xpath('//*[@id="top"]/ul/li['+str(x)+']/div[2]/div[2]/h2/a') 
        mydict['name'] = name_element.text
    
        rating_element = browser.find_elements_by_xpath('//*[@id="top"]/ul/li['+str(x)+']/div[2]/div[2]/div[1]')[0]
        mydict['rating'] = rating_element.text 

        brand_element = browser.find_elements_by_xpath('//*[@id="top"]/ul/li['+str(x)+']/div[2]/div[2]/div[1]/div[1]/a[1]')[0]
        mydict['brand'] = brand_element.text
    
        condition1 = browser.find_elements_by_xpath('//*[@id="top"]/ul/li['+str(x)+']/div[2]/div[2]/div[1]/div[1]/b[2]')[0]
        if condition1.text == "调香师：":
            who_element = browser.find_elements_by_xpath('//*[@id="top"]/ul/li['+str(x)+']/div[2]/div[2]/div[1]/div[1]/a[2]')[0]
            mydict['who'] = who_element.text
            
            condition2 = browser.find_elements_by_xpath('//*[@id="top"]/ul/li['+str(x)+']/div[2]/div[2]/div[1]/div[1]/a[3]')[0]
            blank2 = condition2.text.find(" ")
            if blank2 > 0: 
                gender_element = browser.find_elements_by_xpath('//*[@id="top"]/ul/li['+str(x)+']/div[2]/div[2]/div[1]/div[1]/a[4]')[0]
                mydict['gender'] = gender_element.text
                try:
                    type_element = browser.find_elements_by_xpath('//*[@id="top"]/ul/li['+str(x)+']/div[2]/div[2]/div[1]/div[1]/a[5]')[0]
                    mydict['types'] = type_element.text
                except:
                    mydict['types'] = "None"
            else:
                gender_element = browser.find_elements_by_xpath('//*[@id="top"]/ul/li['+str(x)+']/div[2]/div[2]/div[1]/div[1]/a[3]')[0]
                mydict['gender'] = gender_element.text
                try:
                    type_element = browser.find_elements_by_xpath('//*[@id="top"]/ul/li['+str(x)+']/div[2]/div[2]/div[1]/div[1]/a[4]')[0]
                    mydict['types'] = type_element.text
                except:
                    mydict['types'] = "None"
        else:
            mydict['who'] = "None"
            
            gender_element = browser.find_elements_by_xpath('//*[@id="top"]/ul/li['+str(x)+']/div[2]/div[2]/div[1]/div[1]/a[2]')[0]
            mydict['gender'] = gender_element.text
            
            try:
                type_element = browser.find_elements_by_xpath('//*[@id="top"]/ul/li['+str(x)+']/div[2]/div[2]/div[1]/div[1]/a[3]')[0]
                mydict['types'] = type_element.text
            except:
                mydict['types'] = "None"

        flavor1_element = browser.find_elements_by_xpath('//*[@id="top"]/ul/li['+str(x)+']/div[2]/div[2]/div[1]/div[3]/a[1]/span')[0]
        mydict['scent1'] = flavor1_element.text
    
        flavor2_element = browser.find_elements_by_xpath('//*[@id="top"]/ul/li['+str(x)+']/div[2]/div[2]/div[1]/div[3]/a[2]/span')[0]
        mydict['scent2'] = flavor2_element.text
    
        flavor3_element = browser.find_elements_by_xpath('//*[@id="top"]/ul/li['+str(x)+']/div[2]/div[2]/div[1]/div[3]/a[3]/span')[0]
        mydict['scent3'] = flavor3_element.text
    
        try:
            flavor4_element = browser.find_elements_by_xpath('//*[@id="top"]/ul/li['+str(x)+']/div[2]/div[2]/div[1]/div[3]/a[4]/span')[0]
            mydict['scent4'] = flavor4_element.text
        except:
            mydict['scent4'] = "None"
    
        try:
            flavor5_element = browser.find_elements_by_xpath('//*[@id="top"]/ul/li['+str(x)+']/div[2]/div[2]/div[1]/div[3]/a[5]/span')[0]
            mydict['scent5'] = flavor5_element.text
        except:
            mydict['scent'] = "None"

        result_list.append(mydict.copy()) 
```
