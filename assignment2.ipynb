```python
from math import log
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1

```python
p=0
percentage1=[]
probability1=[]

while True:
    ER = nx.erdos_renyi_graph(10000,p)
    data = ER.degree()
    degree = [item[1] for item in data] 
    iso = degree.count(0)/10000.0
    percentage1.append(iso)
    probability1.append(p)
    p = p+0.0001
    
    if iso == 0:
        break
print percentage1
```

## 2

```python
p = probability1[-1]
print 'times(repetition):',len(percentage1)
print 'probability(no isolated nodes):',p
```

## 3

```python
m = 2*log(10000)/(10000-1)
print 'mathematical result:',m
if p>m:
    print 'p > mathematical result'
else:
    if p<m:
        print 'p < mathematical result'
    else:
        print 'p = mathematical result'
```

```python
# In Erdos-Renyi random graph, n*(1-p)^n = 0, there are no isolated nodes.
# When looking for the threshold 2*ln(n)/(n-1), we assume that n approaches infinity.
# So theoretically, as n increases, p tends to 2*ln(n)/(n-1). In another word, it is the upper bound of p. 
```

## 4

```python
times1 = range(1,len(percentage1)+1)   # [a,b)
plt.plot(times1,percentage1)
plt.xlabel('times')
plt.ylabel('percentage')
```
