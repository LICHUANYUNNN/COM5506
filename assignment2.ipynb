```python
from math import log
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1

```python
p=0
percent1=[]  

while True:
    ER = nx.erdos_renyi_graph(10000,p)
    data = ER.degree()
    degree = [item[1] for item in data] 
    iso = degree.count(0)/10000
    percent1.append(iso)
    p = p+0.0001
    
    if iso <= 0:
        break

print(percent1)
```

## 2

```python
print('times(repetition):',len(percent1))
print('probability(no isolated nodes):',p)
```

## 3

```python
m=2*log(10000)/(10000-1)
print('mathematical result:',m)

if p>m:
    print('p > mathematical result')
else:
    if p<m:
        print('p < mathematical result')
    else:
        print('p = mathematical result')
```

```python
# In Erdos-Renyi random graph, n*(1-p)^n = 0, there are no isolated nodes.

# When looking for the threshold, we assume that n approaches infinity.
# So theoretically, as n increases, p tends to 2*ln(n)/(n-1). In another word, 2*ln(n)/(n-1) is the upper bound of p. 
```

## 4

```python
times1=list(range(1,int(len(percent1)+1)))   # [a,b)
plt.plot(times1,percent1)
plt.xlabel('times')
plt.ylabel('probability')
```
