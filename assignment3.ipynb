```python
import random
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1 Random Graph

```python
n=int(raw_input('What is the size of your network?  n='))
p=float(raw_input('What is the probability of linking two nodes?  p='))
# random.random(): generate a random number between 0 and 1
# undirected graph: i<j

G=nx.Graph()
for i in range(n):
    for j in range(i+1,n):
        if random.random() < p:
            G.add_edge(i,j) 
        else:
            G.add_nodes_from([i,j])

degree=dict(nx.degree(G))
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
nx.draw(G,node_size=10,width=0.5)
plt.subplot(1,2,2)
plt.hist(degree.values(),facecolor='red',alpha=0.5)
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.grid(True)
```

## 2 Small World Model 

```python
p=0
path=[]
cluster=[]

while True:
    WS = nx.watts_strogatz_graph(100,8,p)
    asp = nx.algorithms.average_shortest_path_length(WS)
    ac = nx.average_clustering(WS)
    path.append(asp)
    cluster.append(ac)
    p = p+0.001
    
    if p >= 8.0/(100-1):
        break
```

```python
times = range(1,len(path)+1)  # [a,b)
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(times,path)
plt.xlabel('times')
plt.ylabel('average shortest path')
plt.subplot(1,2,2)
plt.plot(times,cluster)
plt.xlabel('times')
plt.ylabel('average clustering')
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(times,path)
plt.xlabel('times')
plt.ylabel('average shortest path')
plt.xlim(1,50)
plt.subplot(1,2,2)
plt.plot(times,cluster)
plt.xlabel('times')
plt.ylabel('average clustering')
plt.xlim(1,50)
```
