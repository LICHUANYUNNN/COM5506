```python
import random
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n=int(raw_input('What is the size of your network?  n='))

# Scale-free network
# The starting point of this network is 0. At time 0, there is no nodes.
G=nx.Graph()
for i in range(n+1):
    if i > 0:
        G.add_node(i)
        if i > 1:
            j = random.randint(1,i-1)
            G.add_edge(i,j)

# Erdos-Renyi network
ER=nx.erdos_renyi_graph(n,2.0/n)

# Plot
degree1=dict(nx.degree(G))
degree2=dict(nx.degree(ER))
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
nx.draw(G,node_size=[i*5 for i in degree1.values()],width=0.5)
plt.subplot(2,2,2)
plt.hist(degree1.values(),facecolor='red',alpha=0.5)
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.grid(True)
plt.subplot(2,2,3)
nx.draw(ER,node_size=[i*5 for i in degree2.values()],width=0.5)
plt.subplot(2,2,4)
plt.hist(degree2.values(),facecolor='red',alpha=0.5)
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.grid(True)
```
