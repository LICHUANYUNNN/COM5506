```python
import pandas as pd 
import numpy as np 
import seaborn as sns
from pandas import *
from pylab import *
```

```python
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['SimHei'] # 指定默认字体
mpl.rcParams['axes.unicode_minus'] = False   # 解决保存图像是负号'-'显示为方块的问题
```

```python
df = pd.read_csv('perfume_info1.csv', encoding = 'utf8',index_col=0)
```

```python
freq=[]
for i in range(0,len(df)):
    a=df['scent1'][i]+' '+df['scent2'][i]+' '+df['scent3'][i]+' '+df['scent4'][i]+' '+df['scent5'][i]
    freq.append(a)
df['frequency']=freq
```

```python
male=df[df['gender']=='男香']
female=df[df['gender']=='女香']
neu=df[df['gender']=='中性香']
```

```python
male_common = pd.Series(' '.join(male['frequency']).split()).value_counts(ascending=False)
female_common = pd.Series(' '.join(female['frequency']).split()).value_counts(ascending=False)
neu_common = pd.Series(' '.join(neu['frequency']).split()).value_counts(ascending=False)
```

```python
from fontTools.ttLib import TTFont
font = TTFont('D:\Python\SimHei-windows.ttf')
from wordcloud import WordCloud
wordcloud = WordCloud(width = 900, height = 1200, 
                colormap="Blues", background_color ='black',
                font_path="D:\Python\SimHei-windows.ttf",
                max_words=150,
                max_font_size=225, 
                random_state=50,
                min_font_size = 30)
wordcloud.generate_from_frequencies(male_common) 
# plot the WordCloud image                       
plt.figure(figsize = (10,8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
```
